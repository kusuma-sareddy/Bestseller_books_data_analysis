```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import random

# Initialize WebDriver
driver = webdriver.Chrome()
delay = 0.5
# Open the webpage
driver.get("https://www.bookswagon.com/promo-best-seller/best-seller/03AC998EBDC2?sid=560850")
driver.set_window_size(1920, 1080)

# Initialize data lists
more_data = []
category = []
details = []
increment = 3

# Loop through pages
for j in range(1, 451):
    try:
        # Scroll to load content
        count = 0
        while count <= 3:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(delay)
            count += 1

        # Click to load more products
        for i in range(1, increment + 1):
            try:
                driver.find_element(By.XPATH, '//*[@id="lastPostLoader"]/span').click()
#                 sleep(random.uniform(2, 5))
                sleep(delay)
            except Exception as e:
                print(f"Exception in loading more products: {e}")
                continue

        # Increment after every 20 products
        if (j % 20) == 0:
            increment += 1

        # Try clicking the image and handle exceptions
        try:
            driver.find_element(By.XPATH, f'//*[@id="listpromoproduct"]/div[{j}]/div/a/img').click()
            sleep(2)
        except Exception as e:
            print(f"Exception in clicking product image at index {j}: {e}")
            continue

        # Scroll down the page
        driver.execute_script("window.scrollBy(0, 300);")
        sleep(2)

        # Parse the page content
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        text1 = soup.find_all("div", id="bookdetail")
        text2 = soup.find_all("div", class_="col-md-12 mb-3")
        text3 = soup.find_all("div", id="detailcol2")

        for k in text1:
            more_data.append(k.text)
            print(f"{j}  ", k.text)

        for l in text2:
            category.append(l.text)
            print(f"{j}  ", l.text)

        for m in text3:
            details.append(m.text)
#             print(f"{j}  ", m.text)

        # Go back to the previous page
        driver.back()
        sleep(2)

    except Exception as e:
        print(f"Exception in loop at index {j}: {e}")
        continue

# Scroll and click to load more products after the loop
# try:
#     count = 0
#     while count <= 3:
#         driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#         sleep(delay)
#         count += 1

#     for i in range(1, 30):
#         try:
#             driver.find_element(By.XPATH, '//*[@id="lastPostLoader"]/span').click()
#             sleep(6)
#         except Exception as e:
#             print(f"Exception in clicking loader span: {e}")
#             continue

#     # Parse final page content
#     soup = BeautifulSoup(driver.page_source, 'html.parser')
#     a = soup.find_all("div", class_="card align-items-center")

# except Exception as e:
#     print(f"Exception in final scrolling and clicking: {e}")

# Close the driver
driver.quit()
```

```python
No_of_pages=[]
returnable=[]
for s in more_data:
    if re.findall("No\sof\sPages:\s(\d+)",s):
        No_of_pages.append("".join(re.findall("No\sof\sPages:\s(\d+)",s)))
    else:
        No_of_pages.append(np.nan)
    if re.findall("Returnable:\s([A-Z])",s):
        returnable.append("".join(re.findall("Returnable:\s([A-Z])",s)))
    else:
        returnable.append(np.nan)
```

```python

MRP = []
for i in cleaned_details:
    match = re.search(r"(?:\dM\.R\.P\.\s:(₹.*)Save)|(?:Reviews\s+(₹.*)₹)", i)
    
    if match:
        # Get the first non-empty group from the match
        price = match.group(1) if match.group(1) else match.group(2)
        price=price.strip()
        MRP.append(price)
    else:
        MRP.append(np.nan)
```
